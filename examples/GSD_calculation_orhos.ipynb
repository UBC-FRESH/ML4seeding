```python
import rasterio
with rasterio.open("data_set/orthomosaic/G15Segment1.tif") as src:
    print("Pixel size (X, Y):", src.res)
    print("CRS:", src.crs)
```

```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt

with rasterio.open("./data_set/orthomosaic/G15Segment1.tif") as src:
    img = src.read([1,2,3])

img = np.transpose(img, (1,2,0))
plt.imshow(img)
plt.title("Ortho")
plt.show()
```
