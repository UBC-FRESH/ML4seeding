```python
from pathlib import Path
from PIL import Image
import os

# -------- settings --------
BASE_INPUT_DIR = Path("./data_set/raw_drone_images")
BASE_OUTPUT_DIR = Path("./output/compressed_jpgs")
MAX_MB = 20.0

Q_START = 95
Q_MIN = 35
Q_STEP = 5

SCALE_START = 1.00
SCALE_MIN = 0.60
SCALE_STEP = 0.05
# --------------------------

# -------- Folder Selection --------
folders = sorted([
    p.name for p in BASE_INPUT_DIR.iterdir()
    if p.is_dir() and not p.name.startswith(".")
])

if not folders:
    raise FileNotFoundError(f"No valid subfolders found in: {BASE_INPUT_DIR}")

# If dropdown exists, use it; otherwise show numbered selection
if "folder_widget" in globals() and folder_widget is not None:
    SELECTED_FOLDER = folder_widget.value
else:
    print("Available folders:")
    for i, name in enumerate(folders, 1):
        print(f"{i}. {name}")
    choice = int(input(f"\nSelect folder (1-{len(folders)}): "))
    SELECTED_FOLDER = folders[choice - 1]

# Validate selection
if SELECTED_FOLDER not in folders:
    raise ValueError("Invalid folder selected.")

INPUT_DIR = BASE_INPUT_DIR / SELECTED_FOLDER
OUTPUT_DIR = BASE_OUTPUT_DIR / SELECTED_FOLDER
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

print(f"\nProcessing: {SELECTED_FOLDER}")
print(f"Input:  {INPUT_DIR}")
print(f"Output: {OUTPUT_DIR}\n")

# -------- Compression Functions --------
def file_mb(p: Path) -> float:
    return p.stat().st_size / (1024 * 1024)

def save_jpeg(img: Image.Image, out_path: Path, quality: int):
    if img.mode != "RGB":
        img = img.convert("RGB")
    img.save(out_path, format="JPEG", quality=quality, optimize=True, progressive=True)

def compress_to_target(in_path: Path, out_path: Path, max_mb: float = 20.0) -> dict:
    img0 = Image.open(in_path)
    scale = SCALE_START

    while scale >= SCALE_MIN:
        if scale < 1.0:
            w, h = img0.size
            img = img0.resize((int(w * scale), int(h * scale)), Image.Resampling.LANCZOS)
        else:
            img = img0

        q = Q_START
        while q >= Q_MIN:
            save_jpeg(img, out_path, q)
            size = file_mb(out_path)

            if size <= max_mb:
                return {"status": "ok", "quality": q, "scale": scale, "size_mb": round(size, 2)}

            q -= Q_STEP

        scale -= SCALE_STEP

    return {"status": "failed"}

# -------- Run Compression --------
files = sorted([
    p for p in INPUT_DIR.glob("*")
    if p.is_file() and p.suffix.lower() in [".jpg", ".jpeg"]
])
print(f"Found {len(files)} JPGs\n")

for i, f in enumerate(files, 1):
    out = OUTPUT_DIR / f.name
    result = compress_to_target(f, out, MAX_MB)
    print(f"[{i}/{len(files)}] {f.name} -> {result}")
```
