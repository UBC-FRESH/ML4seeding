```python
import os
import shutil

# 🔹 your main folder (images + masks mixed)
input_dir = "./train_a10_2"

# 🔹 output folders
images_dir = os.path.join(input_dir, "images")
masks_dir = os.path.join(input_dir, "masks")

os.makedirs(images_dir, exist_ok=True)
os.makedirs(masks_dir, exist_ok=True)

# 🔹 define how masks are identified
def is_mask(filename):
    return ("mask" in filename.lower()) or ("seg" in filename.lower())

# 🔹 loop through files
for file in os.listdir(input_dir):
    file_path = os.path.join(input_dir, file)

    if os.path.isfile(file_path):
        if is_mask(file):
            shutil.move(file_path, os.path.join(masks_dir, file))
        else:
            shutil.move(file_path, os.path.join(images_dir, file))

print("✅ Done! Files separated into images/ and masks/")
```
