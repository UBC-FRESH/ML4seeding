```python
from pathlib import Path
from PIL import Image

def convert_jpg_to_png(folder):
    folder = Path(folder)

    for jpg_file in folder.glob("*.jpg"):
        img = Image.open(jpg_file).convert("RGB")

        png_file = jpg_file.with_suffix(".png")
        img.save(png_file)

        jpg_file.unlink()  # delete original JPG

    print(f"Converted images in {folder}")

# Example
convert_jpg_to_png("./")
```
